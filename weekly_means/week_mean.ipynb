```python
#Import 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; seaborn.set()
```

```python
col_names = ['Date', 'WaveHeight', 'WavePeriod', 'WaveDir', 'WindSpd', 'WindDir']
file_name = 'test_data.txt'
data = pd.read_csv(file_name, names=col_names, index_col='Date', parse_dates=True)
```

```python
data['WaveHeight'].plot()
plt.ylabel('Hourly Stats')
plt.show
```

```python
data['Year'] = data.index.year
data['Week'] = data.index.week
five_yr = data['2010':'2015']
print five_yr
```

```python
piv = pd.pivot_table(five_yr, index=['Week'], columns=['Year'], values=['WaveHeight'])
print piv 
```

```python
sns.set()
fig, ax = plt.subplots(figsize=(20,10))
pd.pivot_table(five_yr, index=['Week'],columns=['Year'], values=['WaveHeight']).plot(ax=ax, legend=False)

plt.title('Average Weekly Wave Heights')
plt.ylabel('Wave Height (m)')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.savefig('all_years.png')
```
